###Problem-02###

```python
sequences = open("C:/Users/DELL/Downloads/rosalind_gc (1).txt", 'r')

max_gc = 0
max_id = None
filter_fasta = {}
dna_id = None
dna = ""

for line in sequences:
    if line[0] == '>':
        if dna_id is not None:
            filter_fasta[dna_id.replace("\n", "")] = dna.replace("\n", "")
            dna = ""
        dna_id = line[1:].strip()
    else:
        dna += line.strip()
filter_fasta[dna_id.replace("\n", "")] = dna.replace("\n", "")

for fasta in filter_fasta:
    dna_seq = filter_fasta[fasta]
    gc_count = dna_seq.count("G") + dna_seq.count("C")
    gc_percentile = (gc_count / len(dna_seq)) * 100
    if gc_percentile > max_gc:
        max_gc = gc_percentile
        max_id = fasta

print(max_id)
print(round(max_gc, 6))
```
